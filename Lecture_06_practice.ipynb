```python
import pandas as pd
```

### Creating two random dataframes

```python
df_A = pd.DataFrame()
df_A['Name'] = ['Kawsar', 'Tasnim', 'Jakir']
# df_A['name'] = ['Kawsar', 'Tasnim', 'Jakir']
df_A['Address'] = ['Houston', 'Boise', 'Tampa']
df_A['University'] = ['USF', 'BSU', 'USF']

display(df_A)
```

```python
df_B = pd.DataFrame()
df_B['Name'] = ['Kawsar', 'Tasnim', 'Riad']
df_B['Department'] = ['EE', 'CS', 'EE']
df_B['Address'] = ['Texas', 'Idaho', 'Dhaka']

display(df_B)
```

### inner join

```python
df_inner_e1 = df_A.merge(df_B, how='inner', on=['Name'], suffixes=('_L', '_R'))
# df_inner_e1 = df_A.merge(df_B, how='inner')

display(df_inner_e1)
```

```python
# df_inner_e2 = df_B.merge(df_A, how='inner', left_on='Name', right_on='name')

# display(df_inner_e2)
```

```python
df_inner_e1 = df_A.merge(df_B, how='inner')
df_inner_e2 = df_A.merge(df_B, how='inner', on='Name')
df_inner_e3 = df_A.merge(df_B, how='inner', on='Address')

display(df_inner_e1)
display(df_inner_e2)
display(df_inner_e3)
```

### outer join

```python
df_outer_e1 = df_A.merge(df_B, how='outer')
df_outer_e2 = df_A.merge(df_B, how='outer', on='Name')
df_outer_e3 = df_A.merge(df_B, how='outer', on='Address')

display(df_outer_e1)
display(df_outer_e2)
display(df_outer_e3)
```

```python
df_left_e1 = df_A.merge(df_B, how='left')
df_left_e2 = df_A.merge(df_B, how='left', on='Name')
df_left_e3 = df_A.merge(df_B, how='left', on='Address')

display(df_left_e1)
display(df_left_e2)
display(df_left_e3)
```

### left join

```python
# df_left = df_A.merge(df_B, how='left', left_on='name', right_on='Name')

# display(df_left)
```

```python
df_left = df_B.merge(df_A, how='left', on='Name')

display(df_left)
```

### right join

```python
# df_right = df_A.merge(df_B, how='right', left_on='name', right_on='Name')

# display(df_right)
```

```python
df_right = df_B.merge(df_A, how='right', on='Name')

display(df_right)
```

### joining by index

```python
df_A = pd.DataFrame()
df_A['Name'] = ['Kawsar', 'Tasnim', 'Jakir']
df_A['Address'] = ['Houston', 'Boise', 'Tampa']
df_A['University'] = ['USF', 'BSU', 'USF']
df_A = df_A.set_index('Name')

display(df_A)

df_B = pd.DataFrame()
df_B['name'] = ['Kawsar', 'Tasnim', 'Riad']
df_B['Department'] = ['EE', 'CS', 'EE']
df_B = df_B.set_index('name')

display(df_B)
```

```python
df_inner = df_A.merge(df_B, how='inner', left_index=True, right_index=True)

display(df_inner)
```

### Data joining in friends dataset

```python
friends_s1 = pd.read_excel("friends.xlsx", sheet_name='birthday')
friends_s2 = pd.read_excel("friends.xlsx", sheet_name='gender')

display(friends_s1)
display(friends_s2)
```

```python
friends_inner = friends_s1.merge(friends_s2, 
                           how='inner', 
                           left_on='character', 
                           right_on='character_name')

display(friends_inner)
```

```python
friends_outer = friends_s1.merge(friends_s2,
                           how='outer', 
                           left_on='character', 
                           right_on='character_name')

display(friends_outer)
```

```python
friends_left = friends_s1.merge(friends_s2, 
                           how='left', 
                           left_on='character', 
                           right_on='character_name')

display(friends_left)
```

```python
friends_right = friends_s1.merge(friends_s2, 
                           how='right', 
                           left_on='character', 
                           right_on='character_name')

display(friends_right)
```
